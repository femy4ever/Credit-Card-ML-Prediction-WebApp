# Data Cleaning

```python
# checking for duplicates
duplicateCount = creditCard_data.duplicated().sum()
print("Number of duplicate records: ", duplicateCount)
```

```python
# dropping duplicates
creditCard_data = creditCard_data.drop_duplicates(keep='first')
duplicateCount = creditCard_data.duplicated().sum()
print("Number of duplicate records is now: ", duplicateCount)
```

```python
# Checking the number of transaction after droping duplicate
creditCard_data['Class'].value_counts()
```

```python
# checking for outliers in the amount column
sns.boxplot(creditCard_data['Amount'])
```

```python
# Checkig Outlier count
outlier = creditCard_data['Amount'] > 10000
outlier.count()
```

```python
# Using Zero Method to address Outliers
upperLimit = creditCard_data["Amount"].mean() + 3 * creditCard_data["Amount"].std()
lowerLimit = creditCard_data["Amount"].mean() - 3 * creditCard_data["Amount"].std()
print("upperLimit:", upperLimit)
print("lowerLimit:", lowerLimit)
```

```python
# capping method: changing the outlier to upper and lower limit values
data = creditCard_data.copy()
data.loc[(data ["Amount"] > upperLimit), "Amount"] = upperLimit
data.loc[(data ["Amount"] < lowerLimit), "Amount"] = lowerLimit
```

```python
# checking for outliers in our new dataset
sns.boxplot(data['Amount'])
```

```python
# #sns.distplot(data['Amount'])
sns.distplot(data['Amount'])
plt.show()
```

```python
# checking for skewness
data['Amount'].skew()
```

```python
# Reduction of data skewness
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method= 'yeo-johnson')
data['Amount'] = pt.fit_transform(data[['Amount']])
data['Amount'].skew()
```

```python
# #sns.distplot(data['Amount'])
sns.distplot(data['Amount'])
plt.show()
```

```python
## Viewing the cleaned dataset
data.head()
```

```python
# dataset value counts()
data['Class'].value_counts()
```

```python
# using StandardScaler techniques
scaler = StandardScaler()
data['Amount'] = scaler.fit_transform(data['Amount'].values.reshape(-1, 1))
data['Time'] = scaler.fit_transform(data['Time'].values.reshape(-1, 1))
```

```python
data.describe()
```

```python
#feature selection 
new_data = data.copy()

# Calculate correlation with the target feature
correlation_with_target = new_data.drop('Class', axis=1).apply(lambda x: x.corr(new_data['Class']))

# Filter features with low correlation to the target feature
low_correlation_features = correlation_with_target[abs(correlation_with_target) < 0.05].index

# Ensure 'Time' and 'Amount' are not dropped
if 'Time' in low_correlation_features:
    low_correlation_features = low_correlation_features.drop('Time')
if 'Amount' in low_correlation_features:
    low_correlation_features = low_correlation_features.drop('Amount')

# Drop low correlation features from the dataset
new_data_filtered = new_data.drop(low_correlation_features, axis=1)
```

```python
new_data_filtered.describe()
```

```python
new_data_filtered['Class'].value_counts()
```

```python
# Treating imbalanced dataset using Undersampling
non_fraudulent = new_data_filtered[new_data_filtered['Class']==0]
fraud = new_data_filtered[new_data_filtered['Class']==1]
#checking non_fraudulent shape
non_fraudulent.shape
```

```python
non_fraudulent_sample=non_fraudulent.sample(n=473)
non_fraudulent_sample.shape
```

```python
model_data = pd.concat([non_fraudulent_sample,fraud],ignore_index=True)
```

```python
model_data['Class'].value_counts()
```

```python
model_data.head()
```

```python
# Splitting target feature
X = model_data.drop("Class", axis=1)
y = model_data["Class"]
```

```python
# Setting Train and Test Data
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = True, test_size = 0.3, random_state = 42 )

print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of testing label:', y_test.shape)
```

```python
# Model Designing

def evaluate_model(model, X_test, y_test):

    # Predict Test Data 
    y_pred = model.predict(X_test)

    # Calculating accuracy, precision, recall, f1-score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)
    class_report = metrics.classification_report(y_test, y_pred)

    # Calculate area under curve (AUC)
    y_pred_proba = model.predict_proba(X_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(12,5))
    cm = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
    sns.heatmap(cm, annot=True,fmt='d',cmap="Blues");

    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 
            'fpr': fpr, 'tpr': tpr, 'auc': auc, 'cm': cm}
```

```python
#LogisticRegression
lngr = LogisticRegression(random_state=0, max_iter=500, )
lngr.fit(X_train,y_train)
lngr_eval = evaluate_model(lngr, X_test, y_test)

# Print result
print('Accuracy of Logistic Regression:',lngr_eval['acc'])
print('Precision:', lngr_eval['prec'])
print('Recall:',lngr_eval['rec'])
print('F1 Score:',lngr_eval['f1'])
print('Area Under Curve:', lngr_eval['auc'])
print('Confusion Matrix:\n', lngr_eval['cm'])
```

```python
import joblib

# Save the trained model to a file
joblib.dump(lngr, 'lngr_model.pkl')
```

```python
#Random Forest 
rfc = RandomForestClassifier(random_state=0) 
rfc.fit(X_train, y_train)
rfc_eval = evaluate_model(rfc, X_test, y_test)

# Print result
print('Accuracy of Randomforest:', rfc_eval['acc'])
print('Precision:', rfc_eval['prec'])
print('Recall:',rfc_eval['rec'])
print('F1 Score:',rfc_eval['f1'])
print('Area Under Curve:', rfc_eval['auc'])
print('Confusion Matrix:\n', rfc_eval['cm'])
```
